```python
# create a collapsed version of classes

import os

import string
import glob
import random

def maybe_makedir(dirname, force=False):
  if os.path.isdir(dirname) and not force:
    # You may override by setting force=True.
    print('%s already present - Skipping making dir' % (dirname))
  else:
    print('Making dir %s.' % dirname)
    os.makedirs(dirname)
  return 
```

```python
original_classes = '/Volumes/LZD1601/training_data/benthoz2015'
collapsed_classes = '/Volumes/LZD1601/training_data_collapsed/benthoz2015'
balanced_classes = 'Volumes/LZD1601/training_data_balanced/benthoz2015'
```

```python
# number of classes
catalog = list()
for path, dirs, files in os.walk(original_classes):
    
    #print  path, " has ", 
    #print  len(dirs), " dirs ",
    #print  " and ", len(files) , " files"
    if len(dirs)==0:
        catalog.append({'class': os.path.basename(path), 'examples': len(files)})
        print catalog[-1]

```

```python
min_examples = 500
max_examples = 2000
usable_classes = 0
for entry in catalog:
    if entry['examples'] > min_examples:
        print entry
        usable_classes = usable_classes+1
print('classes with more than %i examples: %i') % (min_examples,usable_classes)
```

```python
for path, dirs, files in os.walk(original_classes):
    if len(dirs)==0:
        # no sub directories means it's a class folder with only files
        num_examples = len(files)
        class_name = os.path.basename(path)
        #print  class_name, " has ", 
        #print  " and ", num_examples , " files"
        if num_examples >= min_examples:
            print('using class %s with %i examples') % (class_name, num_examples)
            # can be used
            if num_examples >= max_examples:
                # sumsample before symlink
                symfiles = random.sample(files,max_examples)
            else:
                symfiles = files
            print("Linking %i images") % len(symfiles)
            destination = os.path.join(collapsed_classes,class_name)
            maybe_makedir(destination)
            for entry in symfiles:
                destination_file = os.path.join(destination,entry)
                source_file = os.path.join(path,entry)
                os.symlink(source_file,destination_file)
        else:
            print('* skipping class %s with %i examples') % (class_name, num_examples)
                
            
            
                
        
```
